```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal 
import scipy.io
from scipy.io import wavfile
```

```python
def sound( x, rate=44100, label=''):
    from IPython.display import display, Audio, HTML
    display( HTML( 
    '<style> table, th, td {border: 0px; }</style> <table><tr><td>' + label + 
    '</td><td>' + Audio( x, rate=rate)._repr_html_()[3:] + '</td></tr></table>'
    ))

def equalizer(input_sound, fs, freq, gains):
    gains = 10.**(np.asarray(gain)/10.)
    eq = signal.firwin2(4097, freq,  gains, window="hamming", fs=fs)
    w, EQ = signal.freqz(eq, 1)
    plt.figure(figsize=(8, 3))
    plt.title("designed eq filter")
    plt.plot(w, EQ)
    plt.show()
    result = np.convolve(eq, input_sound)    
    return result
```

#### Load Source File

```python
fs_i, data_i = wavfile.read("src.wav", "r")
sound(data_i, fs_i)
```

#### Load Output File

```python
# create a butter worth filter here to replace the output file for testing
freq = [0, 500, 510, 520, 1700, 1800, 1900, fs_i/2]
gain = [1, 1, -2, 1, 1, 2, 1, 1]
data_o = equalizer(data_i, fs_i, freq, gain) # add a filter
data_o = np.concatenate((np.zeros(int(fs*0.5)),data_o)) # add a 0.1s delay\

sound(data_o, fs_i)
```

#### Measure frequency response

```python
# make two array same length
print(len(data_i), len(data_o))
diff = len(data_o) - len(data_i)
data_i = np.concatenate((data_i, np.zeros(diff)))
print(len(data_i), len(data_o))
```

```python
# fft both
DATA_I = np.fft.fft(data_i)
DATA_O = np.fft.fft(data_o)
fr = DATA_O / DATA_I

magnitude = np.abs(fr)
angle = np.unwrap(np.angle(fr))
plt.subplots(1,2, figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(magnitude)
plt.title("magnitude of frequency response")
plt.subplot(1,2,2)
plt.plot(angle)
plt.title("angle of frequency response")
```

#### Measure Delay

```python
# find time delay
# https://stackoverflow.com/questions/41492882/find-time-shift-of-two-signals-using-cross-correlation
def lag_finder(y1, y2, sr):
    n = len(y1)

    corr = signal.correlate(y2, y1, mode='same') / np.sqrt(signal.correlate(y1, y1, mode='same')[int(n/2)] * signal.correlate(y2, y2, mode='same')[int(n/2)])

    delay_arr = np.linspace(-0.5*n/sr, 0.5*n/sr, n)
    delay = delay_arr[np.argmax(corr)]
    print('y2 is ' + str(delay) + ' behind y1')

    plt.figure()
    plt.plot(delay_arr, corr)
    plt.title('Lag: ' + str(np.round(delay, 3)) + ' s')
    plt.xlabel('Lag')
    plt.ylabel('Correlation coeff')
    plt.show()


lag_finder(data_i, data_o, fs_i)
```
